```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy
```

```python
img = cv.imread('./images.png', 0)
plt.imshow(img, cmap="gray")


plt.show()
```

```python
ret, bn_th = cv.threshold(img, 127, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)
plt.imshow(bn_th, cmap = 'gray')
plt.show()
```

```python
kar = numpy.ones((5,5), numpy.uint8)
print(kar)
```

```python
erosion = cv.erode(bn_th, kar, iterations = 1)
plt.imshow(erosion, cmap = 'gray')
plt.show()
```

```python
dilation = cv.dilate(bn_th, kar, iterations = 1)
plt.imshow(dilation, cmap = 'gray')
plt.show()
```

```python
erodi = cv.dilate(erosion, kar, iterations = 1)
plt.imshow(erodi, cmap = 'gray')
plt.show()
```

```python
open_img = cv.morphologyEx(bn_th, cv.MORPH_OPEN, kar)
plt.imshow(open_img, cmap = 'gray')
plt.show()
```

```python
diero = cv.morphologyEx(bn_th, cv.MORPH_CLOSE, kar)
plt.imshow(diero, cmap = 'gray')
plt.show()
```
